```python

X_mutual = sp_data.dropna().drop(columns=["name", "sector"]).astype("int64").reset_index().copy()
y_mutual = X_mutual.pop("price")
del X_mutual["index"]
discrete_features = X_mutual.dtypes == int
```

```python
def make_mi_score(X_mutual, y_mutual, discrete_features):
    mi_scores = mutual_info_regression(X_mutual, y_mutual, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X_mutual.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_score(X_mutual, y_mutual, discrete_features)
mi_scores.to_frame().T
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.figure(dpi=150, figsize=(13, 5))
    plt.barh(width, scores, color="blue")
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores", 
              weight="bold", 
              fontsize=15, pad=25)
plot_mi_scores(mi_scores)
```

```python
def feature_engineering(data):
    data["52_week_add"] = data["52_week_low"] + data["52_week_high"]
    data["52_week_sub"] = data["52_week_low"] - data["52_week_high"]
    data["52_week_div"] = data["52_week_low"] / data["52_week_high"]
    data["52_week_mul"] = data["52_week_low"] * data["52_week_high"]
    return data
sp_data_feat = feature_engineering(sp_data).copy()
```

```python

set_seed()
X = sp_data.drop(columns=["name", "price"])
y = sp_data["price"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.2, random_state=42
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
set_seed()
numerical_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="mean")),
    ("scaler", MinMaxScaler())
])
categorical_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder())
])
preprocessor = ColumnTransformer([
    ("numeric", numerical_pipe, list(X_train.drop(columns=["sector"]).columns)),
    ("categoric", categorical_pipe, ["sector"])
])
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", KNeighborsRegressor(n_neighbors=1, n_jobs=-1))
])
parameter = {
    "model__n_neighbors": np.arange(1, 10, 2),
    "model__p": [1, 2],
    "model__weights": ["uniform", "distance"]
}
```

```python
set_seed()
model = GridSearchCV(pipeline, param_grid=parameter, cv=3, n_jobs=-1, verbose=1)
model.fit(X_train, y_train)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
set_seed()
numerical_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="mean")),
    ("scaler", MinMaxScaler())
])
categorical_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder())
])
preprocessor = ColumnTransformer([
    ("numeric", numerical_pipe, list(X_train.drop(columns=["sector"]).columns)),
    ("categoric", categorical_pipe, ["sector"])
])
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestRegressor(n_jobs=-1, random_state=42))
])
parameter = {
    "model__n_estimators": [100, 200, 300, 400, 500],
    "model__max_depth": [2, 3, 5, 7, 9],
    "model__max_features": ["auto", "sqrt", "log2"],
    "model__min_samples_leaf": [2, 3, 5, 7, 9]
}
```

```python
set_seed()
model = GridSearchCV(pipeline, param_grid=parameter, cv=3, n_jobs=-1, verbose=0)
model.fit(X_train, y_train)
```

```python
print(f"model train score: {model.score(X_train, y_train)},\n"
      f"model best score: {model.best_score_},\n"
      f"model test score: {model.score(X_test, y_test)},\n"
      f"model best estimator: {model.best_estimator_}")
```
